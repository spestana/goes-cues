Compare the spectral response functions for GOES-16 ABI band 14, ASTER band 14, and MODIS band 31.

```python
import pandas as pd
import numpy as np
import xarray as xr
import matplotlib.pyplot as plt
```

```python
# GOES-16 ABI Band 14
g16abi14_srf = pd.read_csv(r"C:\Users\steve\OneDrive\Documents\School Stuff\UW\Mountain Hydrology Research Group\spectral response functions\goes17abi\fm2\cimss.ssec.wisc.edu\goes\calibration\SRF\abiFM2\ABI_FM2_10Mar2016_CH14.SRF",
                          sep="\s+", header=None, names=['wavenumber', 'response'])
g16abi14_srf['wavelength_um'] = (1/g16abi14_srf.wavenumber)*10000

# Terra MODIS Band 31
modisTerra31_srf = pd.read_csv(r"C:\Users\steve\OneDrive\Documents\School Stuff\UW\Mountain Hydrology Research Group\spectral response functions\modis\rtcoef_eos_1_modis_srf_ch31.txt", 
                                       skiprows=4, sep="\s{3,}", header=None, names=['wavenumber', 'response'])
modisTerra31_srf['wavelength_um'] = (1/modisTerra31_srf.wavenumber)*10000

# Terra ASTER Band 14
aster_srf = pd.read_csv(r"C:\Users\steve\OneDrive\Documents\School Stuff\UW\Mountain Hydrology Research Group\spectral response functions\aster\aster_tir.txt",
                        skiprows=4, sep="\t",header=None, names=['wavelength_10_um', 'response_10', 
                                                                 'wavelength_11_um', 'response_11', 
                                                                 'wavelength_12_um', 'response_12', 
                                                                 'wavelength_13_um', 'response_13',
                                                                 'wavelength_14_um', 'response_14'])
```

```python
fig, ax = plt.subplots(figsize=(10,4))
g16abi14_srf.plot(x='wavelength_um', y='response', label='GOES-16 ABI Band 14', ax=ax, color='k', linestyle='-')
modisTerra31_srf.plot(x='wavelength_um', y='response', label='Terra MODIS Band 31', ax=ax, color='k', linestyle='--')
aster_srf.plot(x='wavelength_14_um', y='response_14', label='Terra ASTER Band 14', ax=ax, color='k', linestyle=':')

ax.set_xlim(10.3, 12.0)
ax.legend(loc='upper left')
```
