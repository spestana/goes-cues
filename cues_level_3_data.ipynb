```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

Open the CUES data file

```python
cues_file_path = r'C:\Users\steve\OneDrive\Documents\School Stuff\UW\Mountain Hydrology Research Group\CUES_data\cues2011to2017_Radiation_Albedo_WindSpeed_GroundTemp_AirTemp_RH_AirPressure_SnowDepth_SWE.csv'
cues_data = pd.read_csv(cues_file_path)
```

Convert the 'DateTime' text strings to pandas datetime objects

```python
cues_data['datetime'] = pd.to_datetime(cues_data['DateTime'])
```

Look at all the keys (column headers)

```python
cues_data.keys()
```

Filter to dates within a certain time window

```python
startdate = pd.datetime(2017,3,1)
enddate = pd.datetime(2017,4,1)
time_window = ( cues_data['datetime'] > startdate ) & ( cues_data['datetime'] < enddate ) 
```

Plot some of these values within this time window

```python
fig, ax = plt.subplots(figsize=(15,7))
cues_data[time_window].plot(x='DateTime',y='uplooking_broadband_watts_per_meterxx2', ax=ax)
cues_data[time_window].plot(x='DateTime',y='uplooking_direct_watts_per_meterxx2', ax=ax)
cues_data[time_window].plot(x='DateTime',y='uplooking_diffuse_watts_per_meterxx2', ax=ax)
cues_data[time_window].plot(x='DateTime',y='uplooking_longwave_watts_per_meterxx2', ax=ax)
```
